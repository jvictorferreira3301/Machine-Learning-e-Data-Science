O Pandas é uma biblioteca em Python para manipulação e análise de dados. É amplamente utilizada por cientistas de dados, analistas e engenheiros para trabalhar com dados tabulares de maneira eficiente e flexível. 

- **Estruturas de Dados Principais:** O Pandas oferece duas estruturas de dados principais:

    - **Series:** Um array unidimensional que pode conter qualquer tipo de dado (inteiros, strings, floats, etc.). Similar a um vetor, mas com rótulos para cada elemento.

    - **DataFrame:** Uma estrutura de dados bidimensional (tabelar) que pode ser vista como uma coleção de Series alinhadas. É semelhante a uma tabela em uma base de dados ou uma planilha Excel, com rótulos (nomes) para linhas e colunas.

- **Leitura e Escrita de Dados:** Pandas pode facilmente ler dados de vários formatos, incluindo CSV, Excel, SQL, JSON, HTML e muitos outros. Ele também pode exportar DataFrames para esses formatos.

- **Manipulação de Dados:** O Pandas oferece uma variedade de métodos para manipular dados, incluindo:

    - Seleção e filtragem de dados.
    - Manipulação de índices.
    - Agrupamento e agregação de dados.
    - Tratamento de dados ausentes.
    - Junção e mesclagem de conjuntos de dados.
    - Alteração de tipos de dados.
    
- **Visualização:** Integrado com bibliotecas como Matplotlib e Seaborn, Pandas permite criar gráficos diretamente a partir de DataFrames para uma rápida análise visual.

### 4. Manipulação de Dados


```python
# Renomear coluna 'final-weight' para 'fnlwgt'
df.rename(columns={'final-weight': 'fnlwgt'}, inplace=True)
df.rename(columns={'fnlwgt': 'final-weight'}, inplace=True)


# Adicionar uma nova coluna com base em uma condição
df['is_adult'] = df['age'] >= 18

# Remover colunas 'capital-gain' e 'capital-loos'
df.drop(columns=['capital-gain', 'capital-loos'], inplace=True)

# Substituir valores na coluna 'income'
df['income'].replace({'<=50K': 'low', '>50K': 'high'}, inplace=True)

# Preencher valores ausentes na coluna 'native-country' com 'Unknown'
df['native-country'].fillna('Unknown', inplace=True)
```

### 5. Análise de Dados


```python
# Agrupar dados pela coluna 'marital-status' e calcular a média de idade
df.groupby('marital-status')['age'].mean()
```

```python

# Agrupar dados por 'education' e calcular a soma de 'hour-per-week'
df.groupby('education')['hour-per-week'].sum()
```

```python

# Ordenar dados pela coluna 'age'
df.sort_values(by='age', ascending=False)
```

```python

# Criar tabela dinâmica para média de horas trabalhadas por semana por sexo e educação
df.pivot_table(values='hour-per-week', index='sex', columns='education', aggfunc='mean')
```

### 6. Visualização de Dados


```python
import matplotlib.pyplot as plt

# Plotar histograma da coluna 'age'
df['age'].hist()
plt.title('Distribuição de Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()

# Plotar gráfico de dispersão entre 'age' e 'hour-per-week'
df.plot.scatter(x='age', y='relationship', title='Idade vs. Horas por Semana')
plt.show()

# Plotar gráfico de barras para contagem de 'education'
df['education'].value_counts().plot.bar()
plt.title('Distribuição de Educação')
plt.xlabel('Educação')
plt.ylabel('Contagem')
plt.show()
```
