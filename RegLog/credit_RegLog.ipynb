## Regressão Logística

A regressão logística é um método estatístico usado para modelar a probabilidade de um evento binário (com duas possíveis saídas) ocorrer. É amplamente utilizada em problemas de classificação binária, onde o objetivo é prever a probabilidade de uma amostra pertencer a uma das duas classes.

### Como Funciona

A regressão logística utiliza a função sigmoide para mapear a saída de uma combinação linear das variáveis independentes (features) para um valor entre 0 e 1, que pode ser interpretado como uma probabilidade.

A fórmula da regressão logística é:

$$
P(Y=1|X) = \sigma(\beta_0 + \beta_1X_1 + \beta_2X_2 + \ldots + \beta_nX_n) 
$$

Onde:


$(\sigma(z) = \frac{1}{1 + e^{-z}})$ é a função sigmoide.

$(\beta_0 )$ é o intercepto.

$(\beta_1,\beta_2, \ldots, \beta_n )$ são os coeficientes das variáveis independentes $( X_1, X_2, \ldots, X_n )$

### Aplicações

- **Classificação Binária**: Determinar se um e-mail é spam ou não, se um cliente vai comprar um produto ou não, etc.
- **Medicina**: Prever a presença ou ausência de uma doença.
- **Finanças**: Avaliar a probabilidade de inadimplência de um empréstimo.

### Vantagens

- **Interpretação**: Os coeficientes da regressão logística podem ser interpretados como a mudança na log-odds de o evento ocorrer para uma unidade de mudança na variável independente.
- **Probabilidades**: Fornece probabilidades que podem ser usadas para tomar decisões baseadas em risco.

### Limitações

- **Linearidade**: Assume uma relação linear entre as variáveis independentes e a log-odds do evento.
- **Multicolinearidade**: Pode ser sensível à multicolinearidade entre as variáveis independentes.




```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sn

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix

with open('../datasets/credit.pkl', 'rb') as f:
    X_credit_train, y_credit_train, X_credit_test, y_credit_test = pickle.load(f)
```

```python
RegLog_credit = LogisticRegression(random_state=1)
RegLog_credit.fit(X_credit_train, y_credit_train)
```

Os valores `RegLog_credit.intercept_` e `RegLog_credit.coef_` na célula abaixo são os parâmetros do modelo de regressão logística treinado, `RegLog_credit` Vamos detalhar o que cada um representa:

- `RegLog_credit.intercept_`
Intercepto: Este valor representa o intercepto $((\beta_0))$ do modelo de regressão logística. É o ponto onde a linha de regressão cruza o eixo y quando todas as variáveis independentes são zero. Em termos de probabilidade, ele influencia a probabilidade base do evento ocorrer quando todas as variáveis independentes são zero.

- `RegLog_credit.coef_`
Coeficientes: Este array contém os coeficientes $((\beta_1, \beta_2, \ldots, \beta_n))$ das variáveis independentes do modelo. Cada coeficiente representa a mudança na log-odds da variável dependente para uma unidade de mudança na variável independente correspondente. Em outras palavras, eles indicam a força e a direção da relação entre cada variável independente e a variável dependente.
Exemplo de Interpretação
Se RegLog_credit.intercept_ for 0.5 e RegLog_credit.coef_ for [0.3, -0.2], isso significa:

O intercepto é 0.5, o que influencia a probabilidade base do evento.
O coeficiente 0.3 para a primeira variável independente indica que, para cada aumento de uma unidade nessa variável, a log-odds do evento aumenta em 0.3.
O coeficiente -0.2 para a segunda variável independente indica que, para cada aumento de uma unidade nessa variável, a log-odds do evento diminui em 0.2.

```python
RegLog_credit.intercept_, RegLog_credit.coef_
```

```python
predict_credit = RegLog_credit.predict(X_credit_test)
predict_credit
```

```python
accuracy_score(y_credit_test, predict_credit) 
```

```python
y_true = y_credit_test # Rótulos verdadeiros
y_pred = predict_credit  # Previsões do modelo

# Calcula a matriz de confusão
cm = confusion_matrix(y_true, y_pred)

# Plota a matriz de confusão como um mapa de calor (heatmap)
plt.figure(figsize=(8, 6))
sn.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.xlabel('Previsão')
plt.ylabel('Rotulo Verdadeiro')
plt.title('Matriz de Confusão')
plt.show()
```

```python
print(classification_report(y_credit_test, predict_credit))
```
